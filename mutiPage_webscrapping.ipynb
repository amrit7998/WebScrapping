```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
c_name = []
c_type = []
age  = []
rating  = []
review_popl = []
h_rate = []
c_rate = []
for j in range(1,501):
    headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}
    url = f"https://www.ambitionbox.com/list-of-companies?page={j}"
    webpage = requests.get(url,headers=headers).text
    page = BeautifulSoup(webpage,'lxml')
    c_list = page.find_all('div',class_='companyCardWrapper')

    str=''
    count=0
    for t in c_list:
        rate = t.find_all('span',class_="companyCardWrapper__ratingValues")
        l = t.find('a', href=lambda href: href and href.startswith("https://www.ambitionbox.com/reviews/"))
        review_popl.append(l.text)
        if len(rate)>0:
            h_rate.append(rate[0].text)
        else:
            h_rate.append("")
        if len(rate)>1:
            c_rate.append(rate[1].text)
        else:
            c_rate.append("")

        c_name.append(t.find('h2').text.strip())
        temp = t.find('span',class_="companyCardWrapper__interLinking").text.strip()
        
        agi=0
        typi=0
        strc=0
        for st in temp:
            if st=='|':
                typi=1
                if count==0:
                    count+=1
                    for j in range(len(str)):
                            if ord(str[j])>=48 and ord(str[j])<=57:
                                c_type.append('')
                                strc=1
                                break
                    if strc==0:
                        c_type.append(str)
                if str.find('years old')>0:
                    agi=1
                    age.append(str.strip())
                str=''
            else:
                str+=st
        if agi==0:
            age.append("")
        if typi==0:
            c_type.append("")
        count=0
        str=''
        rating.append(t.find('span',class_='companyCardWrapper__companyRatingValue').text)
        
        
print(review_popl)
print(len(c_name))
print(c_type)
print(len(age))
print(len(rating))
print(len(h_rate))
print(len(c_rate))
print(c_name)
```

```python
data = {'company':c_name, 'type':c_type, 'rating':rating,'reviewers':review_popl ,'age':age, 'highly_ratedFOR':h_rate, 'critically_ratedFOR':c_rate}
c_data = pd.DataFrame(data=data)
c_data.to_csv('ambitionBox_comLIST_multi.csv')
print(c_data)
```
